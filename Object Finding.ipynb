```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_color = cv2.imread('MainImage.jpg')
plt.imshow(image_color)
```

```python
cv2.imshow("Show Image",image_color)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
image_gray = cv2.cvtColor(image_color,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray')
```

```python
template = cv2.imread("Closeup.PNG",0)
cv2.imshow("template",template)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
w, h = template.shape[::-1]
```

```python
result = cv2.matchTemplate(image_gray,template,cv2.TM_CCOEFF_NORMED)
result.shape
```

```python
cv2.imshow("Matching Result", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(result)
```

```python
top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)
cv2.rectangle(image_color,top_left,bottom_right,(0,0,255), 5)
cv2.imshow("truckfound", image_color)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
